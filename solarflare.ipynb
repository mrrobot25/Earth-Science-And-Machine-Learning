```python
import pandas as pd
import numpy as np
```

```python
with open('/content/flare.data2','r+') as f:
	lines=f.read().splitlines()
	print(lines)
```

```python
data = pd.read_csv("/content/flare.data2",skiprows = 1,header=None, 
                   delimiter = " ",
                   names = ['Code_of_class','Spot_size','Spot_distribution',
                            'Acitvity','Evolution','24_hr_activity',
                            'Historically_comlex','Complex_on_this_pass','Area',
                            'Area_of_largest_spot','C_class_flare',
                            'M_class_flare','X_class_flare'])
```

```python
features = data[['Code_of_class','Spot_size','Spot_distribution',
                            'Acitvity','Evolution',
                            'Historically_comlex','Complex_on_this_pass','Area',
                            'Area_of_largest_spot','C_class_flare',
                            'M_class_flare','X_class_flare']]
```

```python
target = data['24_hr_activity']
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
features['Code_of_class'] = LE.fit_transform(features['Code_of_class'])
features['Spot_size'] = LE.fit_transform(features['Spot_size'])
features['Spot_distribution'] = LE.fit_transform(features['Spot_distribution'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,target,
                                                 test_size = 0.2)
```

```python
print(X_train.shape)
X_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
classifier = SVC(gamma='auto')
classifier.fit(X_train,y_train)
predictions = classifier.predict(X_test)
```

```python
matrix = confusion_matrix(y_test,predictions)
```

```python
matrix
```

```python
score = accuracy_score(predictions,y_test)
score
```
