```python
import pandas as pd
import numpy as np
import folium
from folium import plugins
```

# Read Datasets

```python
fire_archive_M6=pd.read_csv("fire_archive_M6_101673.csv")
fire_archive_V1=pd.read_csv("fire_archive_V1_101674.csv")
fire_nrt_M6=pd.read_csv("fire_nrt_M6_101673.csv")
fire_nrt_V1=pd.read_csv("fire_nrt_V1_101674.csv")
```

# Merge archive and nrt data

```python
merged_data=pd.concat([fire_archive_V1,fire_nrt_V1],sort=True)
```

# Fillter data

```python
filtered_data=merged_data.filter(['latitude','longitude','acq_date','frp'])
filtered_data.head()
```

```python
dec24=filtered_data[filtered_data['acq_date']>='2019-12-24']
dec24.head()
```

# highest recorded fire radiation in a day

```python
heighest_radiation=dec24.sort_values(by='frp',ascending=False).head(10)
heighest_radiation
```

```python
m=folium.Map(location=[-35.0,144],control_scale=True,zoom_start=3)
copy_data=heighest_radiation.copy()
for i in range(0,len(copy_data)):
          
          folium.Marker(location=[copy_data.iloc[i]['latitude'],copy_data.iloc[i]['longitude']],
          popup='frp:'+ str(copy_data.iloc[i]['frp']) + '<br/> date:'+ str(copy_data.iloc[i]['acq_date']),
          
          icon=folium.Icon(color='red',icon='fire',prefix="fa"),
          ).add_to(m)
m
        
          
 
```

```python
new_data = dec24[['acq_date','frp']].set_index('acq_date')
peak_dates = new_data.groupby('acq_date').sum().sort_values(by='frp',ascending=False)
peak_dates.head(10)
```
